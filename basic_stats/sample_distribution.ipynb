# This notebook checks sample distributions


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from random import sample
```

```python
sample_size = 7

x = [np.random.normal(0, 10) for _ in range(500000)]
x_sample = [sample(x, sample_size) for _ in range(10000)]

x_sample_mean = [np.mean(v) for v in x_sample]
x_sample_var = [np.var(v) for v in x_sample]


_, ax = plt.subplots(3, 1, figsize=(20, 10))
ax[0].hist(x, bins=100)
ax[1].hist(x_sample_mean, bins=100)
ax[2].hist(x_sample_var, bins=100)

plt.show()
```
