# This notebook checks sample distributions


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from random import sample
```

```python
sample_size = 7

x = [np.random.normal(0, 10) for _ in range(500000)]
x_sample = [sample(x, sample_size) for _ in range(10000)]

x_sample_mean = [np.mean(v) for v in x_sample]
x_sample_var = [np.var(v) for v in x_sample]


_, ax = plt.subplots(3, 1, figsize=(20, 10))
ax[0].hist(x, bins=100)
ax[1].hist(x_sample_mean, bins=100)
ax[2].hist(x_sample_var, bins=100)

plt.show()
```

# For Bernoulli 

```python
N = 1000000
n = 3000
n_trial = 10000
p = .2
x = [np.random.binomial(1, p) for _ in range(N)]

sample_mean = [np.mean(sample(x, n)) for _ in range(n_trial)]
sample_var =  [np.var(sample(x, n)) for _ in range(n_trial)]
```

```python
z = 1.96
confidence_interval = z * np.sqrt(p*(1 - p)/n)
print(f"confidence interval: +/-{confidence_interval}")
print(f"confidence interval range: [{p-confidence_interval}, {p+confidence_interval}]")
plt.hist(sample_mean, bins=90, normed=True)
plt.show()
```
